## Построение простых графиков

Импортируем библиотеку.

```python
import matplotlib.pyplot as plt
```

Создадим фигуру и оси, используя множественное присваивание. В результате получаем пустой график.

```python
fig, ax  = plt.subplots()
```

Регулируем размер графика.

```python
fig, ax  = plt.subplots(figsize=(2, 2))
```

Рисуем простой график — прямую линию.

```python
import numpy as np
xpoints = np.array([0, 5])
ypoints = np.array([3, 10])

fig, ax  = plt.subplots(figsize=(2, 2))
plt.plot(xpoints, ypoints)
plt.show()
```

Меняем линию на точки.

```python
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Построение диаграммы рассеяния

Диаграмма рассеяния подходит для визуализация совместного распределения двух величин

Чтобы нарисовать диаграмму рассеяния, необходимо задать два вектора, и с помощью функции `scatter()` реализовать совместное распределение.


```python
avg_temp = np.array([20, 18, 19, 22, 24, 25, 26, 23, 21, 20])
num_ice_creams = np.array([100, 80, 90, 120, 150, 160, 170, 130, 110, 100])

plt.scatter(avg_temp, num_ice_creams)
plt.show()
```

Пример наложения диаграмм рассеяния.

```python
# текущий год
avg_temp = np.array([20, 18, 19, 22, 24, 25, 26, 23, 21, 20])
num_ice_creams = np.array([100, 80, 90, 120, 150, 160, 170, 130, 110, 100])
prev_year_avg_temp = np.array([20, 21, 20 , 25, 30, 28, 27, 25, 22, 23])
prev_year_num_ice_creams = np.array([100, 70, 80, 100, 160, 130, 140, 110, 95, 90])

# прошлый год
plt.scatter(avg_temp, num_ice_creams, label='2023')
plt.scatter(prev_year_avg_temp, prev_year_num_ice_creams, c='orange', label='2022')

plt.show()
```

На диаграммах рассеяния мы можем добавлять третьи измерения, не вводя отдельную систему координат.

```python
avg_temp = np.array([20, 18, 19, 22, 24, 25, 26, 23, 21, 20])
num_ice_creams = np.array([100, 90, 95, 120, 150, 160, 170, 130, 110, 100])

# дополнительные данные по выручке
total = np.array([150, 120,100,300,290,300, 320,210,200,300])

plt.scatter(avg_temp, num_ice_creams, c = total, cmap='viridis')
plt.colorbar()
plt.show()
```

## Построение пузырьковой диаграммы

Пузырьковая диаграмма — разновидность диаграммы рассеяния. Размер точки на графике пропорционален значению какой-то третьей переменной.

```python
plt.scatter(avg_temp, num_ice_creams, s=total)
plt.show()
```

## Построение столбчатой диаграммы

Столбчатая диаграмма используется для визуализации совместного распределения числовой и категориальной переменной.

```python
x = np.array(["А", "Б", "В", "Г"])
y = np.array([6, 10, 2, 1])

plt.bar(x,y)
plt.show()
```

Функция `barh()` используется для создания горизонтальной столбчатой диаграммы:

## Построение гистограммы

Гистограмма является одним из базовых средств визуализации, поскольку она представляет распределение числовой переменной.

При представлении гистограммы, как правило, на горизонтальной оси располагаются значения переменной по возрастанию, по вертикальной – частота встречаемости этих значений в выборке.

Представим, что нам нужно визуализировать нормальное распределение роста 250 человек со средним 175 и стандартным отклонением 10:

```python
# 175 - среднее значение
# 10 – стандартное отклонение
# 250 – число наблюдений
x = np.random.normal(175, 10, 250)

plt.hist(x)
plt.show()
```

Видно, что **график приближен к нормальному**.

Параметр `bins` регулирует число столбцов, что влияет на степень детализации графика.

Получаем гистограмму с большей детализацией:

```python
plt.hist(x, bins = 30)
plt.show()
```

## Построение круговой диаграммы

Круговая диаграмма хорошо подходит для отображения частей целого.

## Кастомизация графиков

### Оси
Функция `xlabel()` задает название абсциссе, `ylabel()` — ординате.

`alpha=0.5` задает 50%-ю прозрачность

```python
avg_temp = np.array([20, 18, 19, 22, 24, 25, 26, 23, 21, 20])
num_ice_creams = np.array([100, 90, 95, 120, 150, 160, 170, 130, 110, 100])
total = np.array([150, 120,100,300,290,300, 320,210,200,300])

plt.scatter(avg_temp, num_ice_creams, s=total, alpha=0.5)
plt.xlabel("Температура")
plt.ylabel("Количество проданного мороженого")
plt.show()
```

### Название
Название к графику добавляется с помощью функции `title`:

```python
x = np.random.normal(175, 10, 250)

plt.hist(x)

plt.title('Распределение переменной X')

plt.show()
```

### Цвет
В функцию отрисовки графика необходимо добавить параметр `color`, чтобы сменить цвет, установленный по умолчанию.

```python
avg_temp = np.array([20, 18, 19, 22, 24, 25, 26, 23, 21, 20])
num_ice_creams = np.array([100, 90, 95, 120, 150, 160, 170, 130, 110, 100])

plt.scatter(avg_temp, num_ice_creams, color = 'red')
plt.show()
```

Можно изменить палитру цветов. В параметр `cmap` передаем значения градиента цвета.

```python
plt.scatter(avg_temp, num_ice_creams, c=total, cmap='Greens')

plt.colorbar()

plt.show()
```

В следующем примере мы используем вектор цветов для задания цвета сегментов круговой диаграммы.

```python
values = np.array([20, 35, 20, 15, 10])
categories = ["А", "Б", "В", " Г", 'Д']

mycolors = ["grey", "salmon", "lightgreen", "skyblue", "yellow"]

plt.pie(values, labels = categories, colors = mycolors)
plt.show()
```

### Легенда
Легенду к графику можно добавить с помощью параметра `legend()`.

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), '-b', label='синус')
ax.plot(x, np.cos(x), '--r', label='косинус')
ax.axis('equal')

leg = ax.legend()
```

Мы можем изменить положение легенды на графике:

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), '-b', label='синус')
ax.plot(x, np.cos(x), '--r', label='косинус')
ax.axis('equal')

leg = ax.legend(loc = 'upper left') # легенда будет расположена сверху слева
```

### Сетка
Функция  `grid()` позволяет наложить сетку на наш график:

```python
xpoints = np.array([0, 5])
ypoints = np.array([3, 10])

fig, ax  = plt.subplots(figsize=(6, 4))

plt.plot(xpoints, ypoints)
plt.grid()

plt.show()
```

Можно задавать параметры сетки:

```python
plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.show()
```

## Комплексный пример построения графика

Представим, что мы изучаем взаимосвязь между возрастом человека и частотой прослушивания музыки в контексте ментального здоровья. 

Предположим, мы провели опрос, где собрали необходимые данные по музыкальному потреблению пользователей, их возрасту и показателям ментального здоровья.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mxmh_survey_results.csv")
df.head()
```

Построим самый диаграмму рассеяния, на которой посмотрим на совместное распределение возраста и частоты прослушивания музыки.

```python
fig, ax  = plt.subplots(figsize=(6, 4))

p1 = plt.scatter(df['Age'], df['Hours per day'])

plt.show()
```

Разобьем выборку на три терциля (три трети) по метрике депрессии и визуализируем совместное распределение по каждому терцилю.


```python
df_1 = df[df["Depression"] > df["Depression"].quantile(0.77)]
df_2 = df[df["Depression"] < df["Depression"].quantile(0.33)]
df_3 = df.loc[(df["Depression"] <= df["Depression"].quantile(0.77))
          	& (df["Depression"] >= df["Depression"].quantile(0.33))]


fig, ax  = plt.subplots(figsize=(6, 4))
p1 = plt.scatter(df_1['Age'], df_1['Hours per day'])
p2 = plt.scatter(df_2['Age'], df_2['Hours per day'])
p3 = plt.scatter(df_3['Age'], df_3['Hours per day'])
plt.show()
```

Откалибруем цвета и размеры точек.

```python
fig, ax = plt.subplots(figsize=(6, 4))

p1 = plt.scatter(df_1['Age'], df_1['Hours per day'], c= 'orange', s = 10)
p2 = plt.scatter(df_2['Age'], df_2['Hours per day'], c='green', s = 10)
p3 = plt.scatter(df_3['Age'], df_3['Hours per day'], color = 'grey', s = 10)

plt.show()
```

Точки пользователей среднего терциля среднего цвета визуально перетягивают слишком много внимания: сделаем их более прозрачными.

```python
p1 = plt.scatter(df_1['Age'], df_1['Hours per day'], c= 'orange', s = 10)
p2 = plt.scatter(df_2['Age'], df_2['Hours per day'], c='green', s = 10)
p3 = plt.scatter(df_3['Age'], df_3['Hours per day'], color = 'grey', s = 10, alpha = 0.2)
plt.show()
```

Добавим сетку, подписи осей и название графика, легенду для расшифровки цветов точек.

```python
p1 = plt.scatter(df_1['Age'], df_1['Hours per day'], c= 'orange', s = 10)
p2 = plt.scatter(df_2['Age'], df_2['Hours per day'], c='green', s = 10)
p3 = plt.scatter(df_3['Age'], df_3['Hours per day'], color = 'grey', s = 10, alpha = 0.2)

plt.grid(c = 'black', linestyle = '--', linewidth = 0.1)

plt.title('Взаимосвязь между возрастом \n и интенсивностью прослушивания музыки \n по терцилям шкалы депрессии')

plt.ylabel('Интенсивность прослушивания музыки \n(часов в неделю)')
plt.xlabel('Возраст (лет)')

plt.legend((p1, p2, p3),
       	('Верхний терциль депрессии',
        	'Средний терциль депрессии',
        	'Нижний терциль депрессии'),
       	scatterpoints=1,
       	loc='upper right',
       	ncol=1,
       	fontsize=8)
plt.show()
```
