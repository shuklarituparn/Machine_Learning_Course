# Plots in pandas

![image.png](https://pandas.pydata.org/docs/_images/04_plot_overview.svg)

```python
import pandas as pd
import numpy as np
```

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range("1/1/2000", periods=100))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(100, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
df.head()
```

```python
df['A'].plot()
```

```python
df.plot()
```

```python
df.plot(title="My plot")
```

```python
df.plot(kind='hist', title="My plot")
```

```python
df.plot(kind='hist', alpha=.5, title="My plot")
```

```python
df.plot(kind='density', title="My plot")
```

```python
df.plot(kind='scatter', x='A', y='B', title="My plot")
```

```python
df.plot(kind='hist', title="My plot", subplots=True) #, figsize=(6, 6)
```

```python
df.plot(kind='hist', title="My plot", subplots=True, layout=(2, 3), figsize=(15, 10))
```

```python
df.hist(figsize=(6, 6))
```

```python
df['A'].hist()
```

```python
df.plot.scatter(x="A", y="B", alpha=0.5)
```

```python
df.plot.box() #subplots=True, layout=(2, 2), figsize=(10, 10)
```

```python
df = pd.DataFrame({
    'sales': [3, 2, 3, 9, 10, 6],
    'signups': [5, 5, 6, 12, 14, 13],
    'visits': [20, 42, 28, 62, 81, 50],
    }, index=pd.date_range(start='2018/01/01', end='2018/07/01', freq='M'))
df.plot.area()
```

```python
df.plot.area(stacked=False)
```

```python
df['day'] = [1, 2, 3, 4, 5, 6]
df.plot.area(x='day')
```
